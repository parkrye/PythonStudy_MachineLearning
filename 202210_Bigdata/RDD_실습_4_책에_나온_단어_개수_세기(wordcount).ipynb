```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setMaster("local").setAppName("WordCount")
sc = SparkContext(conf = conf)
```

```python
directory = "C:\\Users\\mhso_lec\\study_notebook\\data"
filename  = "Book"

book = sc.textFile(f"file:///{directory}\\{filename}")
book.collect()[:3]
```

`map`을 활용해서 단어 쪼개기

```python
map_words = book.map(lambda x : x.split())
map_words.collect()[:3]
```

`flatMap`은 시퀀스를 모두 평평하게(`flat`) 펼쳐준다.

```python
words = book.flatMap(lambda x : x.split())
words.collect()[:20]
```

```python
wordCounts = words.countByValue()
wordCounts
```
